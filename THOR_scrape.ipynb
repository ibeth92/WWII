```python
#import dependencies
import pandas as pd
import os
from sqlalchemy import create_engine
```

```python
thor_path = 'Resources/THOR_WWII_DATA.csv'
thor_df = pd.read_csv(thor_path,encoding='latin-1', low_memory=False)
thor_df.head()
```

```python
#drop unnecessary columns
thor_one = thor_df.drop(columns = ['WWII_ID', 'MASTER_INDEX_NUMBER', 'NAF', 'COUNTRY_FLYING_MISSION', 'TGT_COUNTRY_CODE', 'TGT_TYPE', 'TGT_ID', 'TGT_INDUSTRY_CODE'])
thor_one.head()
```

```python
# drop more columns
thor_two = thor_one.drop(columns = ['TGT_INDUSTRY', 'SOURCE_LATITUDE', 'SOURCE_LONGITUDE', 'UNIT_ID', 'MDS', 'MSN_TYPE', 'TGT_PRIORITY', 'TGT_PRIORITY_EXPLANATION'])
thor_two.head()
```

```python
#drop more columns
thor_three = thor_two.drop(columns = ['AC_ATTACKING', 'ALTITUDE', 'ALTITUDE_FEET', 'TAKEOFF_BASE', 'TAKEOFF_COUNTRY', 'TAKEOFF_LATITUDE', 'TAKEOFF_LONGITUDE', 'AC_LOST'])
thor_three.head()
```

```python
#drop more columns
thor_four = thor_three.drop(columns = ['AC_DAMAGED', 'AC_AIRBORNE', 'AC_DROPPING', 'TIME_OVER_TARGET', 'SIGHTING_METHOD_CODE', 'SIGHTING_EXPLANATION', 'BDA', 'CALLSIGN'])
thor_four.head()
```

```python
#drop more columns
thor_five = thor_four.drop(columns = ['ROUNDS_AMMO', 'SPARES_RETURN_AC', 'WX_FAIL_AC', 'MECH_FAIL_AC', 'MISC_FAIL_AC', 'TARGET_COMMENT', 'MISSION_COMMENTS', 'SOURCE'])
thor_five.head()
```

```python
#drop final columns
final_thor = thor_five.drop(columns = ['DATABASE_EDIT_COMMENTS'])
final_thor
```

```python
final_thor.columns
```

```python
thor_scrape = final_thor.rename(columns={"AIRCRAFT_NAME": "Aircraft Name", "NUMBER_OF_HE": "Number of High Explosives", "TYPE_OF_HE": "Type of High Explosive", "LBS_HE": "Weight of High Explosive (lbs)", "TONS_OF_HE": "Weight of High Explosive (tons)", "NUMBER_OF_IC": "Number of Incendiary", "TYPE_OF_IC": "Type of Incendiary", "LBS_IC": "Weight of Incendiary (lbs)", "TONS_OF_IC": "Weight of Incendiary (tons)", "NUMBER_OF_FRAG": "Number of Fragmentation Weapon", "TYPE_OF_FRAG": "Type of Fragmentation Weapon", "LBS_FRAG": "Weight of Fragmentaion Weapon (lbs)", "TONS_OF_FRAG": "Weight of Fragmentation Weapon (tons)", "TOTAL_LBS": "Total Weight (lbs)", "TOTAL_TONS": "Total Weight (tons)"})
```

```python
thor_scrape.to_csv("Resources/THOR_scrape.csv")
```
